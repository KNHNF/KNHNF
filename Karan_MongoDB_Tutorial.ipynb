# Insert (Create) Documents

## You can insert individual documents using collection.insert_one().

## In this example, we're going to create four documents and then insert them all with insert_many().


```python
try:
 result = my_collection.insert_many(recipe_documents)

# return a friendly error if the operation fails
except pymongo.errors.OperationFailure:
  print("An authentication error was received. Are you sure your database user is authorized to perform write operations?")
else:
  inserted_count = len(result.inserted_ids)
  print("I inserted %x documents into collection" %(inserted_count))

  print("\n")
```

# Find (Read) Documents

## Now that we have data in Mongo, we can read it. To retrieve all of *the* data in a collection, we call find() with an empty filter.

```python
result = my_collection.find()

if result:
  for doc in result:
    my_recipe = doc['recipe_name']
    my_ingredient_count = len(doc['ingredients'])
    print("%s has %d ingredients" %(my_recipe, my_ingredient_count))

else:
  print("No documents found.")
```

```python
# You can also have it in list format if you need it

[x['recipe_name'] for x in my_collection.find()]
```

```python
# We can also find a single document. Let's find a document
# that has the string "banana" in the ingredients list.


ingredient = 'avocado'
my_doc = my_collection.find_one({"ingredients":ingredient })

if my_doc is not None:
  print("A recipe which uses %s:" %ingredient)
  print(my_doc.get("recipe_name"))
  print("The ingredients are:")
  for i in my_doc.get("ingredients"):
    print("\t-",i)
else:
  print("I didn't find any recipes that contain %s as an ingredient."%ingredient)
print("\n")
```

```python
# Here we find ALL the recipes that contain a certain ingredient
ingredient = "tomato"
contains_ingredient = list(my_collection.find({"ingredients": ingredient.lower()}))


if contains_ingredient:
  print("Recipe which uses '%s':" %(ingredient))
  for doc in contains_ingredient:
    print("\t-",doc.get("recipe_name"))

else:
  print("I didn't find any recipes that contain '%s' as an ingredient." %ingredient)
```

# Update Documents

## This example shows how to update something in the database.
## I changed "Easy Banana Brownies" to "Ignacio's Banana Brownies"

```python

def find_if_recipe_exists(recipe_name):
  '''
  Find if a recipe exists in the database
  '''
  my_recipe = my_collection.find_one({"recipe_name": recipe_name})
  if my_recipe is not None:
    print("Found: " + my_recipe.get("recipe_name"))
  else:
    print("I didn't find any recipes called '%s'." %recipe_name)

find_if_recipe_exists("Easy Banana Brownies")
find_if_recipe_exists("Ignacio's Banana Brownies")

my_result =  my_collection.update_one({"recipe_name": "Easy Banana Brownies"}, {"$set": { 'recipe_name': "Ignacio's Banana Brownies" }})
print("I updated %x records." %(my_result.modified_count))


find_if_recipe_exists("Easy Banana Brownies")
find_if_recipe_exists("Ignacio's Banana Brownies")



```

# Delete Documents

### As with other CRUD methods, you can delete a single document or all documents that match a specified filter. To delete all of the documents in a collection, pass an empty filter to the delete_many() method.

### In this example, we'll attempt to delete two of the recipes. But one of the recipes has been modified so it will not parse the filter and it will delete only one Recipe


```python


my_result = my_collection.delete_many({ "$or": [{ "recipe_name": "Air Fryer Coconut Shrimp" }, { "recipe_name": "Oatmeal Chocolate Coconut Macaroons" }]})
print("I deleted %x records." %(my_result.deleted_count))
```
